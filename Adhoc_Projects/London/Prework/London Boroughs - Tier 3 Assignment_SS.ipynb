**3. Modeling**

Consider creating a function that will calculate a ratio of house prices, comparing the price of a house in 2018 to the price in 1998.

Consider calling this function create_price_ratio.

You'd want this function to:
1. Take a filter of dfg, specifically where this filter constrains the London_Borough, as an argument. For example, one admissible argument should be: dfg[dfg['London_Borough']=='Camden'].
2. Get the Average Price for that Borough, for the years 1998 and 2018.
4. Calculate the ratio of the Average Price for 1998 divided by the Average Price for 2018.
5. Return that ratio.

Once you've written this function, you ultimately want to use it to iterate through all the unique London_Boroughs and work out the ratio capturing the difference of house prices between 1998 and 2018.

Bear in mind: you don't have to write a function like this if you don't want to. If you can solve the brief otherwise, then great! 

***Hint***: This section should test the skills you acquired in:
- Python Data Science Toolbox - Part One, all modules

```python
def create_price_ratio(x): 
    
    #get the average price for 1998 and 2018
    avg_price_1998 = x[x['Year'] == 1998]['Avg_Price'].mean()
    avg_price_2018 = x[x['Year'] == 2018]['Avg_Price'].mean()
    
    #calculate thee ratio 1998/2018 average price
    ratio = avg_price_1998/avg_price_2018
    
    #round it to two decimals
    rounded_ratio = round(ratio, 2)
    
    #return the result
    return(rounded_ratio)
```

```python
# now lets display all the boroughs

unique_boroughs = clean_prop3['Borough'].unique()

borough_ratios = [] # create empthy list so we can story the tuples (borough, ratio)

# a loop to retrieve the results for all the unique boroughs 
for borough in unique_boroughs:
    borough_filter = clean_prop3[clean_prop3['Borough'] == borough] # subsets for the borough
    ratio = create_price_ratio(borough_filter) # apply the function
    borough_ratios.append((borough, ratio)) #append so we can sort later
    
    # Sort the list of tuples based on the price ratio (second element of each tuple)
borough_ratios_sorted = sorted(borough_ratios, key=lambda x: x[1], reverse = True)

# Print the sorted results
for borough, ratio in borough_ratios_sorted:
    print(f"The price ratio for {borough} is: {ratio}")
```

```python
# Lets print the sorted results with the average prices
for borough, ratio in borough_ratios_sorted:
    borough_filter = clean_prop3[clean_prop3['Borough'] == borough]
    avg_price_1998 = borough_filter[borough_filter['Year'] == 1998]['Avg_Price'].mean()
    avg_price_2018 = borough_filter[borough_filter['Year'] == 2018]['Avg_Price'].mean()
    
    print(f"The price ratio for {borough} is: {ratio}")
    print(f"   Average Price in 1998: £{avg_price_1998:.2f}")
    print(f"   Average Price in 2018: £{avg_price_2018:.2f}")
    print()  # Add a blank line for readability
```

```python
# lets plot the ratio using the borough_ratios_sorted 
# but first we need to combine the tuples

import matplotlib.pyplot as plt
import seaborn as sns

boroughs, ratios = zip(*borough_ratios_sorted) # combine the tuples

# Create a color map that is gradient based on the count of rations using len
colmap2 = sns.cubehelix_palette(len(borough_ratios_sorted), as_cmap=True) 

# Let's Normalize ratios to [0, 1] for gradient mapping
norm_ratios = [(r - min(ratios)) / (max(ratios) - min(ratios)) for r in ratios]



# Create a horizontal bar plot
plt.figure(figsize=(12, 10)) # set the size before plotting
plt.barh(boroughs, ratios, color=colmap2(norm_ratios)) # plots the bar chart
plt.xlabel('Price Ratio')
plt.ylabel('Borough')
plt.title('Price Ratios for London Boroughs')


plt.show()


```
